```python
# Define the range of alpha values, starting slightly above 0 to avoid division by zero issues
alphas = np.linspace(0.1, 1000, 100)  # 100 points from 0.1 to 1000

# Initialize lists to store the results
first_moments = []

# Loop over each alpha value, compute the first moment, and store it
for alpha in alphas:
    r_dist = R(alpha=alpha, beta=1.0, d=3)  # Assuming beta and d are fixed
    first_moments.append(r_dist.first_moment)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(alphas, first_moments, label='First Moment of r', marker='o')
plt.title('First Moment of r vs. Alpha')
plt.xlabel('Alpha')
plt.ylabel('First Moment of r')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from r_dist import R

# Define the range of alpha values, starting slightly above 0 to avoid division by zero issues
alphas = np.linspace(0.1, 1000, 100)  # 100 points from 0.1 to 1000

# Initialize lists to store the results
differences = []

# Loop over each alpha value, compute the difference, and store it
for alpha in alphas:
    r_dist = R(alpha=alpha, beta=1.0, d=3)  # Assuming beta and d are fixed
    differences.append(1 - r_dist.first_moment)

# Plotting on a logarithmic scale
plt.figure(figsize=(10, 6))
plt.plot(alphas, differences, label='1 - First Moment of r')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.title('Difference Between 1 and First Moment of r vs. Alpha on Log Scale')
plt.xlabel('Alpha')
plt.ylabel('1 - First Moment of r')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from r_dist import R

# Define the range of alpha values, starting slightly above 0 to avoid division by zero issues
alphas = np.linspace(0.1, 1000, 100)  # 100 points from 0.1 to 1000

# Initialize lists to store the results
differences = []

# Loop over each alpha value, compute the difference, and store it
for alpha in alphas:
    r_dist = R(alpha=alpha, beta=1.0, d=5)  # Assuming beta and d are fixed
    differences.append(np.abs(1 - r_dist.first_moment))

# Plotting with the y-axis on a logarithmic scale
plt.figure(figsize=(10, 6))
plt.plot(alphas, differences, label='1 - First Moment of r')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Logarithmic Scale of Difference Between 1 and First Moment of r vs. Alpha')
plt.xlabel('Alpha')
plt.ylabel('1 - First Moment of r (Log Scale)')
plt.grid(True)
plt.legend()
plt.show()
```
